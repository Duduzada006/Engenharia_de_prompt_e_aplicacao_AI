```python
# Este script Python simula a verificação da temperatura de um sistema.
# Ele foi projetado para ilustrar como exibir um alerta de superaquecimento
# com base em um valor de temperatura predefinido.

def verificar_temperatura(temperatura_atual):
    """
    Verifica a temperatura atual do sistema.
    Se a temperatura for superior a 30°C, exibe um alerta de superaquecimento.
    Caso contrário, informa que a temperatura está normal.

    Args:
        temperatura_atual (int ou float): O valor da temperatura a ser verificada.
    """
    # Condição para verificar se a temperatura excede o limite de 30°C
    if temperatura_atual > 30:
        # Exibe uma mensagem de alerta se a temperatura estiver alta
        print("ALERTA: Superaquecimento! Temperatura atual: " + str(temperatura_atual) + "°C")
    else:
        # Exibe uma mensagem de temperatura normal se não exceder o limite
        print("Temperatura normal: " + str(temperatura_atual) + "°C")

# --- Exemplos de uso da função 'verificar_temperatura' ---
# Você pode substituir os valores simulados por uma função que obtenha
# a temperatura real do sistema (por exemplo, usando bibliotecas como 'psutil'
# para sistemas operacionais específicos, ou sensores de hardware).

# Exemplo 1: Teste com uma temperatura abaixo do limite (temperatura normal)
temperatura_simulada_1 = 28
print("Testando com temperatura normal:")
verificar_temperatura(temperatura_simulada_1)

# Exemplo 2: Teste com uma temperatura acima do limite (superaquecimento)
temperatura_simulada_2 = 35
print("\nTestando com temperatura de superaquecimento:")
verificar_temperatura(temperatura_simulada_2)

# Exemplo 3: Teste com uma temperatura exatamente no limite
temperatura_simulada_3 = 30
print("\nTestando com temperatura no limite:")
verificar_temperatura(temperatura_simulada_3)

# --- Exemplo Interativo: Digite a temperatura ---
print("\n--- Teste Interativo de Temperatura ---")
try:
    temperatura_digitada = float(input("Por favor, digite a temperatura atual para verificação: "))
    verificar_temperatura(temperatura_digitada)
except ValueError:
    print("Entrada inválida. Por favor, digite um número para a temperatura.")
```

### Organizador de Arquivos por Extensão

Este programa Python ajuda a organizar arquivos em um diretório, movendo-os para subpastas criadas com base em suas extensões. Por exemplo, todos os arquivos `.pdf` serão movidos para uma pasta `PDF`.

**Como usar:**
1.  **`diretorio_origem`**: O diretório onde os arquivos bagunçados estão.
2.  **`diretorio_destino`**: O diretório onde as novas pastas organizadas serão criadas. Se for `None`, as pastas serão criadas dentro do `diretorio_origem`.

```python
import os
import shutil

def organizar_arquivos(diretorio_origem, diretorio_destino=None):
    """
    Organiza arquivos em um diretório com base em suas extensões.

    Cria subpastas para cada tipo de arquivo (ex: 'PDF', 'JPG')
    e move os arquivos para suas respectivas pastas.

    Args:
        diretorio_origem (str): O caminho para o diretório que contém os arquivos a serem organizados.
        diretorio_destino (str, optional): O caminho para o diretório de destino onde
                                         as pastas organizadas serão criadas. Se None,
                                         usa o diretorio_origem como destino.
    """
    # Verifica se o diretório de origem existe
    if not os.path.isdir(diretorio_origem):
        print(f"Erro: O diretório de origem '{diretorio_origem}' não existe.")
        return

    # Cria o diretório de destino se especificado e não existir
    if diretorio_destino and not os.path.isdir(diretorio_destino):
        os.makedirs(diretorio_destino)
        print(f"Diretório de destino '{diretorio_destino}' criado.")

    # Define o diretório onde as pastas organizadas serão criadas
    diretorio_alvo = diretorio_destino if diretorio_destino else diretorio_origem

    print(f"Iniciando organização de arquivos em '{diretorio_origem}'...")

    # Itera sobre todos os itens no diretório de origem
    for nome_arquivo in os.listdir(diretorio_origem):
        caminho_arquivo = os.path.join(diretorio_origem, nome_arquivo)

        # Ignora se o item for um diretório
        if os.path.isdir(caminho_arquivo):
            continue

        # Obtém a extensão do arquivo (ex: .pdf, .jpg)
        _, extensao_arquivo = os.path.splitext(nome_arquivo)
        extensao_arquivo = extensao_arquivo.lower() # Converte para minúsculas para padronização

        # Define o nome da pasta com base na extensão
        if not extensao_arquivo: # Caso o arquivo não tenha extensão
            nome_pasta_extensao = "SEM_EXTENSAO"
        else:
            # Remove o '.' da extensão e converte para maiúsculas para o nome da pasta
            nome_pasta_extensao = extensao_arquivo[1:].upper()

        # Cria o caminho completo para a pasta de destino da extensão
        pasta_destino = os.path.join(diretorio_alvo, nome_pasta_extensao)

        # Cria a pasta de extensão se ela não existir
        if not os.path.exists(pasta_destino):
            os.makedirs(pasta_destino)
            print(f"Pasta '{nome_pasta_extensao}' criada em '{diretorio_alvo}'.")

        # Move o arquivo para a pasta de sua extensão
        try:
            shutil.move(caminho_arquivo, os.path.join(pasta_destino, nome_arquivo))
            print(f"Movido: '{nome_arquivo}' para '{nome_pasta_extensao}/'")
        except Exception as e:
            print(f"Erro ao mover '{nome_arquivo}': {e}")

    print("Organização de arquivos concluída!")

# --- Exemplo de Uso ---
```
