#*Calculadora simples*

```python
def add(x, y):
    # Esta função retorna a soma de x e y.
    return x + y

def subtract(x, y):
    # Esta função retorna a diferença entre x e y.
    return x - y
```

Agora, vamos testar as funções de adição e subtração:

```python
num1 = 10
num2 = 5

# Exibe o resultado da soma de num1 e num2
print(f'{num1} + {num2} = {add(num1, num2)}')
# Exibe o resultado da subtração de num2 de num1
print(f'{num1} - {num2} = {subtract(num1, num2)}')
```
